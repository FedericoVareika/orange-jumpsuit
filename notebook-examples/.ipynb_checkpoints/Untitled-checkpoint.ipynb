```python
import time
import numpy as np
import pandas as pd
import faiss
from datasets import load_dataset
from sentence_transformers import SentenceTransformer

# Import your custom library
from orange_jumpsuit import IndexPQ as JumpsuitPQ

# ==========================================
# 1. LOAD AND PREPARE DATASET
# ==========================================
print("Loading GitHub issues dataset...")
# Load the exact dataset used in the HF Chapter 5 tutorial
issues_dataset = load_dataset("lewtun/github-issues", split="train")

# Filter out pull requests and rows without comments
issues_dataset = issues_dataset.filter(
    lambda x: (x["is_pull_request"] == False and len(x["comments"]) > 0)
)

# For a quick benchmark, let's take a subset (e.g., 2,000 issues)
# Remove the .select() call to benchmark the entire dataset
if (len(issues_dataset) > 2000): 
    issues_dataset = issues_dataset.select(range(2000))

# Concatenate title and body to form the search corpus
corpus = [
    f"{title} \n {body}" 
    for title, body in zip(issues_dataset["title"], issues_dataset["body"])
]
```

```python
# ==========================================
# 2. CREATE TEXT EMBEDDINGS
# ==========================================
print("\nEncoding dataset (this may take a moment depending on your GPU/CPU)...")
# Using the model recommended in the Hugging Face tutorial
model = SentenceTransformer("sentence-transformers/multi-qa-mpnet-base-dot-v1")

# Create dense float32 arrays 
corpus_embeddings = model.encode(corpus, convert_to_numpy=True).astype(np.float32)

# Define our query
query = "How can I load a dataset offline?"
print(f"\nEncoding query: '{query}'")
query_embedding = model.encode([query], convert_to_numpy=True).astype(np.float32)
```

```python
# ==========================================
# 3. SET UP INDEX PARAMETERS
# ==========================================
d = corpus_embeddings.shape[1]  # 768 for the mpnet model
m = 8                           # Number of subvectors (768 / 8 = 96)
nbits = 8                       # 256 centroids per subvector
k = 3                           # Retrieve top 3 results

print(f"Dimension: {d}, Base Vectors: {len(corpus_embeddings)}")
```

```python
# ==========================================
# 4. FAISS BENCHMARK
# ==========================================
print("\n--- FAISS Natural Language Search ---")
# HF tutorial defaults to IndexFlatIP. We use IndexPQ to ensure an apples-to-apples 
# architectural comparison with Jumpsuit.
faiss_index = faiss.IndexPQ(d, m, nbits)

t0 = time.time()
faiss_index.train(corpus_embeddings) 
faiss_index.add(corpus_embeddings)
faiss_build_time = time.time() - t0

t0 = time.time()
faiss_distances, faiss_indices = faiss_index.search(query_embedding, k)
faiss_search_time = time.time() - t0
```

```python
# ==========================================
# 5. JUMPSUIT BENCHMARK
# ==========================================
print("\n--- Jumpsuit Natural Language Search ---")
jumpsuit_index = JumpsuitPQ(dimension=d, n_subvectors=m, n_bits_per_value=nbits)

t0 = time.time()
jumpsuit_index.train(corpus_embeddings)
jumpsuit_index.add(corpus_embeddings)
jump_build_time = time.time() - t0

t0 = time.time()
jump_distances, jump_indices = jumpsuit_index.search(query_embedding, k)
jump_search_time = time.time() - t0
```

```python
# ==========================================
# 6. EVALUATION & DISPLAY
# ==========================================
results = {
    "Metric": ["Train & Add Time (s)", "Search Time (s)"],
    "FAISS": [faiss_build_time, faiss_search_time],
    "Jumpsuit": [jump_build_time, jump_search_time]
}
df = pd.DataFrame(results).set_index("Metric")
print("\n=== PERFORMANCE COMPARISON ===")
display(df.round(4))

print("\n=== SEARCH RESULTS FOR FAISS ===")
for idx, dist in zip(faiss_indices[0], faiss_distances[0]):
    print(f"[Score: {dist:.4f}] Title: {issues_dataset[int(idx)]['title']}")

print("\n=== SEARCH RESULTS FOR JUMPSUIT ===")
for idx, dist in zip(jump_indices[0], jump_distances[0]):
    print(f"[Score: {dist:.4f}] Title: {issues_dataset[int(idx)]['title']}")
```
